### Multiple Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mtcars.csv')
```

```python
df
```

```python
df.corrwith(df['mpg'])
```

```python
salary = pd.read_csv('Salary_Data.csv')
salary.corr()
```

```python
x = df[['disp','hp','wt']]
y = df['mpg']
```

```python
x
```

#### Import the class

```python
from sklearn.linear_model import LinearRegression
```

#### Create the object

```python
reg = LinearRegression()
```

#### Train the model

```python
reg.fit(x,y)
```

#### Product on unseen data

```python
disp = 221
hp = 115
wt = 2.51
reg.predict([[disp,hp,wt]])
```

```python
reg.score(x,y)
```

#### Visualize

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.scatter(df['disp'],y)
plt.subplot(2,2,2)
plt.scatter(df['hp'],y)
plt.subplot(2,2,3)
plt.scatter(df['wt'],y)
```

### Decison Tree Classification

#### Import the dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.shape
```

```python
df
```

#### Separate input and output variables

```python
x = df[['Age','EstimatedSalary']]
y = df['Purchased']
```

```python
x
```

```python
y.value_counts()
```

#### Visualize the data

```python
plt.figure(figsize=(10,5))
plt.title('Purchase Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'],x['EstimatedSalary'],c=y)
```

#### Import the data

```python
from sklearn.tree import DecisionTreeClassifier
```

#### Create the object

```python
classifier = DecisionTreeClassifier(random_state=0)
```

#### Train the model

```python
classifier.fit(x,y)
```

#### Plot the tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plit
```

```python
plt.figure(figsize=(12,15))
plot_tree (classifier, fontsize=7, feature_names=['age','sal'],class_names=['y','n'],filled=True, rounded=True);
```

#### Predict on the new value

```python
new=  pd.DataFrame({
    'Age':[23,45,54,67],
    'EstimatedSalary':[45000,89000,23000,134000]
})
```

```python
new
```

```python
classifier.predict(new)
```
